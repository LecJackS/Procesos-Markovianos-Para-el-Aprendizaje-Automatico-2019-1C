This notebook has pieces of code to make it easier the process of converting several pngs to one animated gif.

Is necessary hardcode the name's list to have them in proper order (by number, and not as imagemagick defaults, which has some windows 95 order: 01 10 11 12 ... 02 20 21 22 ... 03 30 31 32 ... etc.

After creating the .gif with:

    command = 'convert -delay 20 plot-snap-*.png animated.gif'
    command = 'convert -delay 20 plot-snap-01.png plot-snap-02.png ... plot-snap-50000.png animated.gif'
    
Uploaded it to giphy.com and downloaded the compressed version for this notebook.

Much more work can be done into this pretty akward (and slow) process, but it'll be fine if it's just a couple of animations.

### Para los archivos de la carpeta self-learn

```python
# For converting images in right order
from subprocess import call
import pprint
# Suppose your files are like 1.jpeg, 2.jpeg etc. upto 100.jpeg

files = []
folder='self-learn'
filename='self-learn-value'

file_dir = folder+'/'+filename
nSnaps = 70
for i in range(0, nSnaps):
    files.append(file_dir+'-{}.png'.format(i))


command = ('convert -delay 4 {} '+folder+'/'+filename+'-'+str(nSnaps)+'-iter.gif').format(' '.join(files))
pprint.pprint(command)
#command = 'convert -delay 20 plot_step_00*.png animated.gif'

call(command, shell=True)
```
