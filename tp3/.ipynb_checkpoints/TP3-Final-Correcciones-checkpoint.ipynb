```python
import gym
import numpy as np
import random
from gym import spaces
from gym.utils import seeding
from collections import OrderedDict
```

## El juego

Reglas:

- 2 jugadores (uno sos vos y una IA)
- Cada jugador cuenta con **5 monedas** (cara o cruz). Cada jugador solo ve sus monedas. 
- Después de tirar las monedas, empieza el juego.
- El juego consiste en intentar adivinar al menos **cuantas caras hay en total** entre todas las monedas o en hacer que el **rival adivine incorrectamente**.
- La apuesta empieza en 0 caras.
- Se tira una moneda para ver quien de los dos jugadores empieza. 
- Acciones posibles:
    * me quedo con la apuesta hasta ahora y paso.
    * subo en 1 la cantidad de caras que creo que hay en total.
- El juego termina cuando algún jugador pasa.
- Si la apuesta fue **mayor** al numero real de caras, **el apostador pierde** (r=-1) y el otro gana.
- Si la apuesta es **menor o igual** al numero real, **gana el apostador** (r=1) y pierde el otro.
- IA inicial de la computadora: Si la apuesta es menor a 2 + cantidad de caras propias, la computadora apuesta. Sino pasa.

```python
# Defino nuestro propio environment con las reglas del juego

def throw_coin(num_coin, np_random):
    return np_random.rand(num_coin)>0.5

def total_faces(list_players):
    RV=0
    for player in list_players:
        RV += sum(player)
    return RV

class PerudoSimplificado(gym.Env):
    def __init__(self):
        self.action_space = spaces.Discrete(2)
        self.observation_space = spaces.Tuple((
            spaces.Discrete(5),   # mis monedas
            spaces.Discrete(10))) # apuesta actual

        self.seed()
        # Empieza el juego
        self.reset()

    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]

    def step(self, action):
        assert self.action_space.contains(action)
        
        max_guess_player_2 = total_faces([self.player_2]) + int(len(self.player_1)/2)
        faces_tot = total_faces([self.player_1, self.player_2])
        
        if self.guess > len(self.player_1) + len(self.player_2): #la apuesta es mayor que lo maximo posible
            done = True
            reward = -1

        if action == 0: #action == 0, mantengo la apuesta y paso
            done = True
            if self.guess <= faces_tot: #tenia razon el rival
                reward = -1
            else: #tenia razon yo
                reward = +1
            
        else: #action == 1, subo en 1 la apuesta
            self.guess += 1
            if self.guess < max_guess_player_2: #el rival sube 1 en su turno
                self.guess += 1 
                done = False
                reward = 0
            else: # el rival pasa
                done = True
                if self.guess <= faces_tot: 
                    reward = 1
                else:
                    reward = -1
            
        return self.get_obs(), reward, done, {}

    def get_obs(self):
        return (sum(self.player_1), self.guess)
    
    def reset(self):
        self.player_1 = throw_coin(5,self.np_random)
        self.player_2 = throw_coin(5,self.np_random)
        self.guess = 1 if np.random.rand()>0.5 else 0 #al principio se tira una moneda para ver quien empieza. Por eso es 0 (uno empieza) o 1 (empezo el otro y subio)  
        return self.get_obs()
```

## Jugando al azar

```python
env = PerudoSimplificado()
print(env.observation_space)
print(env.action_space)
```

```python
#Politica Random:
for i_episode in range(1):
    state = env.reset()
    print("Starting at state:", state)
    while True:
        action = env.action_space.sample() # Selecciona una accion random dentro de las posibles
        #action = 1
        state, reward, done, info = env.step(action) # Juega un paso
        print(state, action)
        if done:
            print('El juego ha terminado! Su reward: ', reward)
            print('Ganaste :)\n') if reward > 0 else print('Perdiste :(\n')
            break
```

```python
def plot_policy(policy, ascii_pol=False):
    """
    Realiza un grafico en 3 dimensiones de la función valor.
    
    El parametro policy es un diccionario cuyas keys son pares "caras en mi mano" 
    y "apuesta actual" y el valor es la acción a realizar.
    """
    def get_Z(x, y):
        if (x,y) in policy:
            if np.isscalar(policy[x,y]):
                # Action to take
                return policy[x,y]
            else:
                # Policy of kind: [actions_probabilities at state (x,y)]
                return random_argmax(policy[x,y])
        else:
            # Para visualizar que la policy no tiene asignada una acción para dicho estado
            return 25 

    def get_figure(ax, ascii_pol):
        x_range = np.arange(0, 6)
        y_range = np.arange(0, 11)
        X, Y = np.meshgrid(x_range, y_range)
        Z = np.array([[get_Z(x,y) for x in x_range] for y in y_range])
        if ascii_pol:
            # Print ascii policy to copy-paste it
            print(Z)
        surf = ax.imshow(np.flip(Z,0), cmap=plt.get_cmap('Pastel2', 3), vmin=0, vmax=2, extent=[-0.5, 5.5, -0.5, 10.5])
        plt.xticks(x_range)
        plt.yticks(y_range)
        plt.gca().invert_yaxis()
        ax.set_xlabel('caras en mi mano')
        ax.set_ylabel('apuestas')
        ax.grid(color='w', linestyle='-', linewidth=1)
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.1)
        cbar = plt.colorbar(surf, ticks=[0,1,2], cax=cax)
        cbar.ax.set_yticklabels(['0 (pasar)','1 (subir)', 'desconocido'])            
    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_subplot(111)
    get_figure(ax, ascii_pol)
    plt.show()
```

#### On-policy first-visit Monte Carlo control

```python
# Filling these ones was freaking fun!
def get_probs(Q_s, epsilon, nA): 
    # Devuelve un array con las probas de cada accion, para un estado dado
    # policy:   {(s,s): [r_a1, r_a2]}
    # policy_s: [r_a1, r_a2]
    policy_s = np.zeros(nA)
    A_star = random_argmax(Q_s)
    # One-hot vector for A_star
    indicator = np.zeros(nA)
    indicator[A_star] = 1
    policy_s = epsilon/nA + indicator*(1-epsilon)
    return policy_s
```

```python
def generate_episode_from_Q(env, Q, epsilon, nA):
    # episode = {(s, s):[action k taken, inmediate reward_a=k]}
    # with order
    episode = OrderedDict()
    state = env.reset()
    actions = np.arange(nA)
    while True:
        # Generate entire episode (until terminal state)
        action = np.random.choice(actions, p=get_probs(Q[state], epsilon, nA))
        state, reward, done, info = env.step(action) # Juega un paso
        if state not in episode:
            # First Visit MC.
            episode[state] = [action, reward]
        if done:
            # Juego terminado
            break
    return episode
```

```python
def update_Q(env, episode, Q, alpha, gamma, MeanReturns): #Returns):
    nA=env.action_space.n
    use_cumulative_average = True # 
    G=0
    for state in reversed(episode):
        # Adding episode values to 'list' of past Returns
        action_taken  = episode[state][0]
        action_reward = episode[state][1]
        # Using cumulative average for faster results,
        # and memory don't blow up with large number of episodes
        # MeanReturns[state] : [[mean_r_a0, times], [[mean_r_a1, times]]
        action_mean   = MeanReturns[state][action_taken][0]
        action_count  = MeanReturns[state][action_taken][1]
        G = G*gamma + action_reward
        if action_count == 0:
            new_mean = action_reward
        else:
            new_mean = (action_mean * action_count + G)/(action_count+1)
        #print("new mean:",new_mean)
        #updated_returns = [new_mean, action_count+1]
        MeanReturns[state][action_taken] = [new_mean, action_count+1]#updated_returns
        #print("MeanReturns[",state,"][",action_taken,"]: ",MeanReturns[state][action_taken] )
        # Now Q(s,a) takes the mean of Returns(s,a);
        Q[state] = np.take(MeanReturns[state], [0,2]) #[0,2] indexes of actions' mean returns
        #print("Q[state]:",Q[state])
    return Q
```

```python
# Simple test empírico para asegurar que el error que arrastra 
# el mean acumulativo sea despreciable:
def test_cumulative_mean_error():
    # Cumulative Mean vs Mean error to check correct usage
    size = 5*10**6
    # Array of irrational numbers
    values = [np.pi,]*size
    cumu_mean = np.pi

    for i in range(1, size):
        cumu_mean = (cumu_mean*i + values[i])/(i+1)

    print("Array of size", size, ":")
    print("+ Numpy pi - Numpy mean :", np.pi - np.mean(values))
    print("Error after", size, "iterations:")
    print("+ Numpy pi - Cumulative mean:", np.pi - cumu_mean)
    print("Relavive error:", (np.pi - np.mean(values))/(np.pi - cumu_mean))
    # cummulative produces an error greater than the actual mean,
    # but still is 1*10**-9, which is pretty good for our case
test_cumulative_mean_error()
```

#### Useful functions

```python
def random_argmax(v):
    """Like np.argmax(), but if there are several "best" actions,
       chooses and returns one randomly"""
    arguments = np.argwhere(v == np.amax(v)).ravel()
    return np.random.choice(arguments)

def softmax_vect(v):
    """Returns a vector with all softmax values"""
    exp_vect = np.exp(v) # element wise exp()
    return exp_vect/np.sum(exp_vect)
```

```python
def initialize_dict(init_values):
    # Initialize dict for all possible states
    # { (sx,sy): [default_value] } for all key states (sx,sy)
    my_faces = np.arange(0, 6)
    bets = np.arange(0, 11)
    states = np.array(np.meshgrid(my_faces, bets)).T.reshape(-1,2)
    keystates = tuple(map(tuple, states))
    return {k: init_values.copy() for k in keystates}
```

```python
def mc_control(env, num_episodes, alpha, gamma=1.0, eps_start=1.0, eps_decay=.99999, eps_min=0.05):
    """ Q es un diccionario cuyas keys son pares 
            ("caras en mi mano","apuesta actual")       (state) 
            [y el valor de CADA , acción en su posicion](action-value)
        Q{(st, ate): [v_a0, v_a1]} """
    nA = env.action_space.n
    # Initialize Q arbitrarily
    init_values = [-1,]*nA
    Q=initialize_dict(init_values)
    # Initialize Returns (only keeps means and action count)
    init_values = [[0, 0],]*nA
    MeanReturns = initialize_dict(init_values)
    # Empty policy
    policy = {}
    
    epsilon = eps_start
    # Start playin episodes
    for iteration in range(num_episodes):
        # Show progress
        if iteration%(num_episodes/20)==0:
            print("iteration", iteration, "-", round(iteration/num_episodes*100,2), "% completed...", end="\r")
        # epsilon decay
        epsilon = max(epsilon*eps_decay, eps_min)
        # Generate one episode from start to terminal
        episode = generate_episode_from_Q(env, Q, epsilon, nA)
        Q = update_Q(env, episode, Q, alpha, gamma, MeanReturns).copy()
        for state in episode:
            policy[state] = get_probs(Q[state], epsilon, nA)
        # To save snap, copy code here
    # Training completed
    print("iteration", iteration+1, "- Done."+" "*20) # clean last line with \r
    # policy is epsilon-soft at training, so we take argmax of probabilities
    policy_MC = dict((k, random_argmax(v)) for k, v in policy.items())
    return policy_MC, Q
```

### Sarsa - On-policy TD control

```python
def update_Q_sarsa(alpha, gamma, Q, state, action, reward, next_state=None, next_action=None):
    """Returns the updated value of Q after one step on episode"""
    return alpha*(reward + gamma*Q[next_state][next_action] - Q[state][action])
    
def epsilon_greedy(Q, state, nA, eps):
    if np.random.rand()>eps:
        # Act greedly wrt Q with probability 1-epsilon
        # random_argmax: breaks ties randomly
        return random_argmax(Q[state])
    else:
        # Choose randomly with probability epsilon
        return np.random.choice(np.arange(nA))
```

```python
def sarsa(env, num_episodes, alpha=0.1, gamma=1.0, epsmin=0.01, alpha_decay_pow=0.6):
    nA=env.action_space.n
    # I want eps_decay^num_episodes to be close to epsmin:
    # num_episodes * log(eps_decay) = log(epsmin)
    # log(eps_decay) = log(epsmin)/num_episodes
    # eps_decay = e^log(epsmin)/num_episodes
    eps_decay = np.exp(np.log(epsmin)/num_episodes)
    eps=0.1
    init_values = [-1,]*nA
    Q=initialize_dict(init_values)
    # Training starts
    for iteration in range(num_episodes):
        # Show progress
        if iteration%(num_episodes/20)==0:
            print("iteration", iteration, "-", round(iteration/num_episodes*100,2), "% completed...", end="\r")
        # epsilon decay for less exploration over episodes
        eps = max(eps*eps_decay, epsmin)
        # Learning rate decay
        alpha = 1/np.power((iteration+1), alpha_decay_pow)
        # epsilon decay for less exploration over episodes
        # epsilon = 1/t^.1 keep values between -3 and 3 for 100-500K episodes
        # eps = 1/((iteration+1)**eps_pow)
        # Start episode
        state = env.reset()
        action = epsilon_greedy(Q, state, nA, eps)
        while True:
            # Generate entire episode (until terminal state)
            prev_state = state
            state, reward, done, info = env.step(action)
            next_action = epsilon_greedy(Q, state, nA, eps)
            # Update Q[s][a] with Sarsa update 
            Q[prev_state][action] += update_Q_sarsa(alpha, gamma, Q,
                                            prev_state, action, reward,
                                            state, next_action)
            # Keeping same next_action as BEFORE Q update
            action = next_action
            # To save snap, copy code here
            if done:
                # Juego terminado
                break
    # Training completed
    print("iteration", iteration+1, "- Done."+" "*20) # clean last line with \r
    return Q
```

### Q-learning - Off-policy TD control

```python
def update_Q_sarsamax(alpha, gamma, Q, state, action, reward, next_state=None):
    """Returns the updated value of Q after one step on episode"""
    return alpha*(reward + gamma*np.amax(Q[next_state]) - Q[state][action])
```

```python
def q_learning(env, num_episodes, alpha=0.1, gamma=1.0, epsmin=0.01, alpha_decay_pow=0.6):
    nA=env.action_space.n
    # I want eps_decay^num_episodes to be close to epsmin:
    # num_episodes * log(eps_decay) = log(epsmin)
    # log(eps_decay) = log(epsmin)/num_episodes
    # eps_decay = e^log(epsmin)/num_episodes
    eps_decay = np.exp(np.log(epsmin)/num_episodes)
    eps=0.1
    # Initialize Q
    init_values = [-1,]*nA
    Q=initialize_dict(init_values)
    # Training starts
    for iteration in range(num_episodes):
        # Show progress
        if iteration%(num_episodes/20)==0:
            print("iteration", iteration, "-", round(iteration/num_episodes*100,2), "% completed...", end="\r")
        # epsilon decay for less exploration over episodes
        eps = max(eps*eps_decay, epsmin)
        # Learning rate decay
        alpha = 1/np.power((iteration+1), alpha_decay_pow)
        # epsilon decay for less exploration over episodes
        # epsilon = 1/t^.1 keep values between -3 and 3 for 100-500K episodes
        # eps = 1/((iteration+1)**eps_pow)
        # Start episode
        state = env.reset()
        while True:
            # Generate entire episode (until terminal state)
            action = epsilon_greedy(Q, state, nA, eps)
            prev_state = state
            state, reward, done, info = env.step(action)
            # Update Q[s][a] with sarsamax update (q-learning update)
            Q[prev_state][action] += update_Q_sarsamax(alpha, gamma, Q,
                                            prev_state, action, reward,
                                            state)
            # To save snap, copy code here
            if done:
                # Juego terminado
                break
    # Training completed
    print("iteration", iteration+1, "- Done."+" "*20) # clean last line with \r
    return Q
```

# . Smart Environment

```python
# Given code with some mods, to include a settable policy
# Default policy is random
# If bot starts first turn > chooses first action from policy instead of always 1
# If player 1 starts > same as before, vs bot with set policy

def throw_coin(num_coin, np_random):
    return np_random.rand(num_coin)>0.5

def total_faces(list_players):
    RV=0
    for player in list_players:
        RV += sum(player)
    return RV

class PerudoSimplificadoSmart(gym.Env):
    def __init__(self, policy={}):
        self.action_space = spaces.Discrete(2)
        self.observation_space = spaces.Tuple((
                spaces.Discrete(5),   # mis monedas
                spaces.Discrete(10))) # apuesta actual
        #TODO
        #random policy DONE
        #always one policy
        #copy cat policy
        self.policy = policy
        self.seed()
        # Empieza el juego
        self.reset()
        #

    def seed(self, seed=None):
        self.np_random, seed = seeding.np_random(seed)
        return [seed]
    
    def player1_turn(self, action):
        actual_total_faces = total_faces([self.player_1, self.player_2])
        reward = 0
        done=False
        # Player 1 chose action:
        if action == 0:
            # action == 0, mantengo la apuesta y paso (termina)
            done = True
            if self.guess <= actual_total_faces: #tenia razon el rival
                reward = -1
            else: #tenia razon yo
                reward = 1
        else:
            # action == 1, subo en 1 la apuesta
            self.guess += 1
            if self.max_bet_reached():
                # Player one decides to bet for 11, and loses 
                self.guess = 10 # self.guess must be in [0,10]
                done = True
                reward = -1
        return reward, done
    
    def player2_turn(self):
        # Player 2 (bot) chooses action from input policy[state]
        bot_state = self.get_bot_obs()
        if bot_state in self.policy:
            bot_action = self.policy[bot_state]
        else:
            bot_action = np.random.choice(self.action_space.n)
        actual_total_faces = total_faces([self.player_1, self.player_2])
        if bot_action == 0:
            # Bot mantiene la apuesta y pasa (termina)
            done = True
            if self.guess <= actual_total_faces: 
                reward = 1 # Bot passes and lose
            else:
                reward = -1 # Bot passes and win
        else:
            # Bot increases bet
            self.guess += 1
            if self.max_bet_reached():
                # Bot decides to bet for 11, and loses 
                self.guess = 10 # self.guess must be in [0,10]
                done = True
                reward = 1 # Player 1 wins
                return reward, done
            done = False
            reward = 0
        return reward, done
    
    def step(self, action):
        # Game dynamics/rules
        assert self.action_space.contains(action)
        
        # Player_1: User
        reward, done = self.player1_turn(action)
        if done:
            return self.get_obs(), reward, done, {}
        
        # Player_2: Bot using inpuut policy
        reward, done = self.player2_turn()
        return self.get_obs(), reward, done, {}
        
    def get_obs(self):
        return (sum(self.player_1), self.guess)
        
    def get_bot_obs(self):
        return (sum(self.player_2), self.guess)
    
    def max_bet_reached(self):
        return self.guess > (len(self.player_1) + len(self.player_2))
    
    def reset(self):
        # Al principio se tira una moneda para ver quien empieza.
        # Por eso es 0 (uno empieza) o 1 (empezo el otro y subio)  
        self.player_1 = throw_coin(5,self.np_random)
        self.player_2 = throw_coin(5,self.np_random)
        if np.random.rand()>0.5:
            # Bot starts and takes action based on policy
            self.guess = 0
            self.player2_turn()
        else:
            # We start, we still need to make an action
            self.guess = 0
        return self.get_obs()
```

Sabemos que los algoritmos vistos convergen a la política óptima.

En el juego analizado en este TP, esta política óptima dependerá de cómo "juega" el otro jugador (parte del environment).

Por ejemplo, si siempre responde con ceros, nunca podremos alcanzar estados más avanzados del juego, y nuestra política óptima no generalizará al "mismo" environment con mejores estrategias de juego.

En esta sección se analiza muy brevemente qué clase de policy obtenemos si:

1. El environment comienza siguiendo una política de "siempre pasar" (elige siempre cero, solo si empieza apuesta)
2. Se aplica Q-Learning para encontrar una policy óptima para este environment
3. Se actualiza el environment con esta nueva policy encontrada
4. Repetir desde 2.

Se observó que es posible encontrar políticas similares a las óptimas encontradas anteriormente.

Ésto genera varias interrogantes:

* Convergerá a una única política? Será óptima?
* Generalizará a problemas con espacios más grandes?
* Cuál debería ser la relación del tamaño de los intervalos de tiempo entrenamiento y update del environment? Podrán optimizarse como parámetros?

```python
def save_value_plot(V, idx, filenames, vmin=-1, vmax=1):
    """
    Realiza un grafico en 3 dimensiones de la función valor.
        
    El parametro V es un diccionario cuyas keys son pares "caras en mi mano" 
    y "apuesta actual".
    
    idx es un indice que debe aumentar en cada llamada
    """
    def get_Z(x, y):
        if (x,y) in V:
            return V[x,y]
        else:
            return 0

    def get_figure(ax):
        x_range = np.arange(0, 6)
        y_range = np.arange(0, 11)
        X, Y = np.meshgrid(x_range, y_range)
        Z = np.array([get_Z(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))]).reshape(X.shape)
        # Points at front ejes x and y to see translations
        ax.scatter(X[Z!=0], 0, zs=0, zdir='z', s=5, c='k', depthshade=True)
        ax.scatter(0, Y[Z!=0], zs=0, zdir='z', s=5, c='k', depthshade=True)
        #
        surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm)#, vmin=-1, vmax=1)
        ax.set_xlabel('caras en tu mano')
        ax.set_ylabel('apuestas')
        ax.set_zlabel('value')
        ax.view_init(ax.elev, -120)
    fig = plt.figure(figsize=(20, 20))
    ax = fig.add_subplot(211, projection='3d')
    # preparing for animation later
    # Placement 0, 0 would be the bottom left, 1, 1 would be the top right.
    ax.text2D(0.05, 0.95, 'iter: '+str(idx), transform=ax.transAxes, size=72)
    get_figure(ax)
    filename='animations/'+str(filenames)+'/'+str(filenames)+"-value-"+str(idx)+'.png'
    plt.savefig(filename, dpi=48, bbox_inches='tight')
    plt.close(fig)
    plt.clf()
```

```python
def save_policy_plot(policy, idx, filenames):
    """
    Realiza un grafico en 3 dimensiones de la función valor.
    
    El parametro policy es un diccionario cuyas keys son pares "caras en mi mano" 
    y "apuesta actual" y el valor es la acción a realizar.
    """
    def get_Z(x, y):
        if (x,y) in policy:
            if np.isscalar(policy[x,y]):
                # Action to take
                return policy[x,y]
            else:
                # Policy of kind: [actions_probabilities at state (x,y)]
                return random_argmax(policy[x,y])
        else:
            # Para visualizar que la policy no tiene asignada una acción para dicho estado
            return 25 

    def get_figure(ax):
        x_range = np.arange(0, 6)
        y_range = np.arange(0, 11)
        # flip y data 
        y_range = np.flip(y_range)
        X, Y = np.meshgrid(x_range, y_range)
        Z = np.array([[get_Z(x,y) for x in x_range] for y in y_range])
        #print(Z)
        surf = ax.imshow(np.flip(Z,0), cmap=plt.get_cmap('Pastel2', 3), vmin=0, vmax=2, extent=[-0.5, 5.5, -0.5, 10.5])
        plt.xticks(x_range)
        plt.yticks(y_range)
        plt.yticks(y_range, y_range[::-1])
        plt.gca().invert_yaxis()
        ax.set_xlabel('caras en mi mano')
        ax.set_ylabel('apuestas')
        ax.grid(color='w', linestyle='-', linewidth=1)
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.1)
        cbar = plt.colorbar(surf, ticks=[], cax=cax)
        #cbar.ax.set_yticklabels(['0 (pasar)','','1 (subir)', 'desconocido',''])
        #ax.set_ylim(ax.get_ylim()[::-1])
        #cbar.set_ticks([0,0.1,1,1.9,2])
        #cbar.ax.set_yticklabels(['','0 (pasar)','1 (subir)', 'desconocido',''])
    fig = plt.figure(figsize=(5, 5))
    ax = fig.add_subplot(111)
    get_figure(ax)
    # preparing for animation later
    # Placement 0, 0 would be the bottom left, 1, 1 would be the top right.
    ax.text(0.00, 1.05, 'iter: '+str(idx), transform=ax.transAxes, size=18)
    get_figure(ax)
    filename='animations/'+str(filenames)+'/'+str(filenames)+"-policy-"+str(idx)+'.png'
    plt.savefig(filename, dpi=96, bbox_inches='tight')
    plt.close(fig)
    plt.clf()
    
```
